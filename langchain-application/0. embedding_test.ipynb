```python
from openai import OpenAI
client = OpenAI()
```

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0.0
    
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
king_embedding_response = client.embeddings.create(
    input="king",
    model="text-embedding-3-large"
)

king_vector = np.array(king_embedding_response.data[0].embedding)

queen_embedding_response = client.embeddings.create(
    input="queen",
    model="text-embedding-3-large"
)

queen_vector = np.array(queen_embedding_response.data[0].embedding)

cosine_similarity(king_vector, queen_vector)
```

```python
korean_king_embedding_response = client.embeddings.create(
    input="왕",
    model="text-embedding-3-large"
)

korean_king_vector = np.array(korean_king_embedding_response.data[0].embedding)

korean_queen_embedding_response = client.embeddings.create(
    input="여왕",
    model="text-embedding-3-large"
)

korean_queen_vector = np.array(korean_queen_embedding_response.data[0].embedding)

cosine_similarity(korean_king_vector, korean_queen_vector)
```
