```python
from langchain_ollama import ChatOllama

# ollama pull deepseek-r1:1.5b
#llm = ChatOllama(model="deepseek-r1:1.5b")
llm = ChatOllama(model="llama3.2:1b")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

food_prompt = PromptTemplate(
    template="Guess the food based on following information, {information}, Please return the name of Food only",
    input_variables=["information"]
)

response = llm.invoke(food_prompt.invoke({"information": "It's spicy and famous in korea"}))

output_parser = StrOutputParser()
print(output_parser.invoke(response))

food_chain = food_prompt | llm | StrOutputParser()
food_chain.invoke("This food is spicy and famous in korea")
```

```python
from langchain_core.prompts import ChatPromptTemplate

recipe_prompt = ChatPromptTemplate.from_messages([
    ('system', '''Provide the recipe of the food that the user wants, Please return the recipe only as a numbered list'''),
    ('human', 'Can you give me the recipe for making {food}?')
])

recipe_chain = recipe_prompt | llm | StrOutputParser()
recipe_chain.invoke({"food" : "Kimchi stew"})
```

```python
from langchain_core.runnables import RunnablePassthrough

final_chain = {"food" : food_chain} | recipe_chain
final_chain.invoke("This food is spicy and famous in america")
```
